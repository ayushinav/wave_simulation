```julia
using LinearAlgebra
using Plots
using BenchmarkTools
using ProgressMeter
```

```julia
include("Operators.jl");
```

```julia
gaussian(x, a, μ, σ)=  a*exp(- (x- μ)^2/2σ^2);
```

```julia
global const ESCAPE_AXIS= 2; # for 2 dim case, excluding the y-axis
```

```julia
# Research about the appropriate grid size for inhomogeneous media

# Appropriate values
zgrid= 1:100; # m
ygrid= 1:1; # m
xgrid= 1:200; # m
nz, ny, nx= length.([zgrid, ygrid, xgrid]);
```

```julia
# Change tgrid according to dx for inhomogenous media
dx= step(xgrid);
dy= step(ygrid);
dz= step(zgrid);

κ= fill(100., nz, ny, nx); # homogeneous medium with κ= 100 m2/s
κ[25:35, :, :].= 400.; # Adding a conductor (more diffusion)

dt= dx^2/ maximum(κ)* 0.1;
tgrid= 0:dt:1.5;
nt= length(tgrid);
```

```julia
heatmap(xgrid, zgrid, κ[:,1,:], yflip= true, title= "κ", xlabel="x", ylabel= "z", c=:viridis)
```

```julia
T= zeros(nz, ny, nx, nt);


d2T1= zeros(nz, ny, nx);
d2T2= zeros(nz, ny, nx);
d2T3= zeros(nz, ny, nx);

dT1= zeros(nz, ny, nx);
dT2= zeros(nz, ny, nx);
dT3= zeros(nz, ny, nx);

dκ1= zeros(nz, ny, nx);
dκ2= zeros(nz, ny, nx);
dκ3= zeros(nz, ny, nx);
```

$\partial_t T=  \nabla \eta \nabla T$ </br>
$\eta$= thermal diffusivity

```julia
# T = nx ny nz nt

function update_T!(T, d2T1, d2T2, d2T3, dT1, dT2, dT3, η, dη1, dη2, dη3, nt, Δx, Δy, Δz, Δt)
    d2_x!(d2T1, T[:,:,:,nt-1], Δx)
    d2_y!(d2T2, T[:,:,:,nt-1], Δy)
    d2_z!(d2T3, T[:,:,:,nt-1], Δz)
    
    d_x!(dT1, T[:,:,:,nt-1], Δx)
    d_y!(dT2, T[:,:,:,nt-1], Δy)
    d_z!(dT3, T[:,:,:,nt-1], Δz)
    
    d_x!(dη1, η[:,:,:], Δx)
    d_y!(dη2, η[:,:,:], Δy)
    d_z!(dη3, η[:,:,:], Δz)
    
    T[:,:,:, nt].= T[:,:,:, nt-1] + Δt.*η.*(d2T1 .+ d2T2 .+d2T3)+ Δt.* (dη1.*dT1.+ dη2.*dT2.+ dη3.*dT3);
end
```

```julia
# Adding a source at z= 0
T[:,1,1,1]= gaussian.(1:nz, 0.1, 30, 5).+ gaussian.(1:nz, 0.1, 70, 5);
```

```julia
@showprogress for it in 2:nt
    update_T!(T, d2T1, d2T2, d2T3, dT1, dT2, dT3, κ, dκ1, dκ2, dκ3, it, dx, dy, dz, dt);
end
```

```julia
anim_heat= @animate for it in 1:20:nt
    heatmap(xgrid, zgrid, T[:,1,:,it], yflip= true, title= "$it", clim= (0.,0.06), xlabel= "x", ylabel= "z", c=:jet)
end
```

```julia
gif(anim_heat, "heat_diffusion.gif")
```

There are two gaussian heat sources centered at x= 30 and x= 60. </br>
The source at x=30 diffuses further because of being centered on a conductor while the other heat source does not diffuse as much because of entirely being on the medium of lower diffusivity
